# Lab Task: Student Record Management System

In this lab task, we will create a Python program that allows an instructor
to store and manage student records using a dictionary data structure.

The program will:
1. Take student data as input (ID, Name, City)
2. Store data in three separate lists
3. Convert lists into a dictionary
4. Allow the instructor to search student records by ID

## Step 1: Taking Input from Instructor

In this step:
- We will repeatedly ask for student ID, name, and city.
- The data will be stored in three separate lists.
- If the instructor types "stop", the loop will terminate.

```python
# Create empty lists to store student data
student_ids = []
student_names = []
student_cities = []
```

```python
# Repeatedly ask instructor for input
while True:
    student_id = input("Enter Student ID (or type 'stop' to finish): ")

    if student_id == "stop":
        break

    name = input("Enter Student Name: ")
    city = input("Enter Residential City: ")

    # Store data in separate lists
    student_ids.append(student_id)
    student_names.append(name)
    student_cities.append(city)
```

## Step 2: Creating Dictionary from Lists

Now we will:
- Traverse all three lists
- Create a dictionary
- Use Student ID as the key
- Store Name and City as a list value

Format:
{ ID : [Name, City] }

```python
# Create dictionary using the three lists
students_dict = {}
for i in range(len(student_ids)):
    students_dict[student_ids[i]] = [student_names[i], student_cities[i]]

# Display the created dictionary
print("\nStudent Dictionary Created Successfully!")
print(students_dict)
```

## Step 3: Searching Student Record

After creating the dictionary:
- The instructor can search for a student by entering Student ID.
- If ID exists, Name and City will be displayed.
- If ID does not exist, a message will be shown.

```python
# Query the dictionary
while True:
    query_id = input("\nEnter Student ID to search (or type 'stop' to exit): ")

    if query_id.lower() == "stop":
        print("Exiting search.")
        break

    if query_id in students_dict:
        print("Name:", students_dict[query_id][0])
        print("City:", students_dict[query_id][1])
    else:
        print("Student ID not found.")
```
